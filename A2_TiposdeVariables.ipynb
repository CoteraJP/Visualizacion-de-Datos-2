# 4 - Describe el conjunto de datos. ¿Qué datos contiene?
Esta base de datos contiene 119,389 observaciones para un hotel urbano y uno turístico (de los cuales la ubicación no está especificada) desde el año 2015 hasta el 2017. Dentro de la base de datos se ubican infomación acerca de los clientes (pais de origen, sexo, correo, nombre), información acerca de las reservaciones (tipo de reservación, fecha de llegada, fecha de salida, cancelaciones) y el ADR.
