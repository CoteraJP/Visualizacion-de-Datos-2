```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import pandas as pd
#pd.set_option('display.max_columns', 100)

# Importar los datos
df = pd.read_csv("datasets/antropometria-dataset.csv")
df.head()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(x = df["peso"], bins = 10)
plt.title("Histograma Peso")

plt.subplot(1,2,2)
plt.hist(x = df["cintura"], bins = 10)
plt.title("Histograma Cintura")
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df["peso"], df["cintura"])
plt.xlabel("Cintura de persona")
plt.ylabel("Peso de persona")
plt.title("Diagrama de dispersión de Peso vs. Cintura")

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(x = df["sistol"], bins = 10)
plt.title("Histograma Sistol")

plt.subplot(1,2,2)
plt.hist(x = df["diastol"], bins = 10)
plt.title("Histograma Diastol")
```

*Histogrma diastol: asimétricos con sesgo a la derecha*

```python
df_hombre = df[df['sexo'] == 1]
df_hombre.shape
```

```python
df_mujer = df[df['sexo'] == 2]
df_mujer.shape
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(x = df_hombre["sistol"], bins = 10)
plt.title("Histograma Sistol Hombre")

plt.subplot(1,2,2)
plt.hist(x = df_hombre["diastol"], bins = 10)
plt.title("Histograma Diastol Hombre")
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(x = df_hombre["peso"], bins = 10)
plt.title("Histograma Peso Hombre")

plt.subplot(1,2,2)
plt.hist(x = df_hombre["cintura"], bins = 10)
plt.title("Histograma Cintura Hombre")
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df_hombre["peso"], df_hombre["cintura"])
plt.xlabel("Cintura de persona")
plt.ylabel("Peso de persona")
plt.title("Diagrama de dispersión de Peso vs. Cintura Hombre")

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(x = df_mujer["sistol"], bins = 10)
plt.title("Histograma Sistol Mujer")

plt.subplot(1,2,2)
plt.hist(x = df_mujer["diastol"], bins = 10)
plt.title("Histograma Diastol Mujer")
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(x = df_mujer["peso"], bins = 10)
plt.title("Histograma Peso Mujer")

plt.subplot(1,2,2)
plt.hist(x = df_mujer["cintura"], bins = 10)
plt.title("Histograma Cintura Mujer")
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df_mujer["peso"], df_mujer["cintura"])
plt.xlabel("Cintura de persona")
plt.ylabel("Peso de persona")
plt.title("Diagrama de dispersión de Peso vs. Cintura Mujer")

plt.show()
```
