```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("datasets/antropometria-dataset.csv")
df.head()
```

2. Escribe el código en Python para generar box plot, line plot, histogramas de las variables peso, cintura, etc. 

# Boxplots

```python
sns.boxplot(data=df.loc[:, ["peso", "cintura", "cadera", "talla", "edad"]])

plt.show()
```

# Line Plot

```python
sns.lineplot(data=df.loc[:,"peso"])

plt.show()
```

```python
sns.lineplot(data=df.loc[:,"cintura"])

plt.show()
```

```python
sns.lineplot(data=df.loc[:,"cadera"])

plt.show()
```

```python
sns.lineplot(data=df.loc[:,"talla"])

plt.show()
```

# Histogramas

```python
sns.histplot(df["peso"], 
             kde = True).set_title("Histograma de Peso")
plt.show()
```

```python
sns.histplot(df["cintura"], 
             kde = True).set_title("Histograma de Cintura")
plt.show()
```

```python
sns.histplot(df["cadera"], 
             kde = True).set_title("Histograma de Cadera")
plt.show()
```

```python
sns.histplot(df["talla"], 
             kde = True).set_title("Histograma de Talla")
plt.show()
```

```python
sns.histplot(df["edad"], 
             kde = True).set_title("Histograma de Edad")
plt.show()
```
